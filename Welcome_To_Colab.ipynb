```
#Question 1: A shopkeeper wants to calculate total bill after discount.
price = float(input("Enter price: "))
discount_pct = float(input("Enter discount percentage: "))
discount_amount = price * (discount_pct / 100)
final_price = price - discount_amount
print(f"Final price after discount: {final_price:.2f}")
```

```
#Question 2: Check whether a person is eligible to vote.
age = int(input("Enter age: "))
if age >= 18:
    print("Eligible")
else:
    print("Not Eligible")

```

```
#Question 3: Calculate electricity bill based on units consumed.
units = int(input("Enter units: "))
if units <= 100:
    amt = units * 4.5
elif units <= 300:
    amt = 100 * 4.5 + (units - 100) * 7
else:
    amt = 100 * 4.5 + 200 * 7 + (units - 300) * 8
surcharge = amt * 0.20
total_bill = amt + surcharge
print(f"Total bill amount: {total_bill:.2f}")
```
