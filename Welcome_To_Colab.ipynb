```
#Question 8: Calculate electricity bill based on units consumed.
units = int(input("Enter units: "))
slabs = [(100, 4.5), (200, 7), (float('inf'), 8)]
amt = 0
remaining = units
for limit, rate in slabs:
    if remaining <= 0: break
    consumed = min(remaining, limit)
    amt += consumed * rate
    remaining -= consumed
surcharge = amt * 0.20
print(f"Total bill amount: {amt + surcharge:.2f}")
```

```
#Question 20: Create a function to calculate simple interest.
def simple_interest(p, r, t):
    annual_interest = p * (r / 100)
    return annual_interest * t

p = float(input("Enter P: "))
r = float(input("Enter R: "))
t = float(input("Enter T: "))
print(f"Simple Interest: {simple_interest(p, r, t):.2f}")
```

```
#Question 21: A shopkeeper wants to calculate total bill after discount.
price = float(input("Enter price: "))
discount_pct = float(input("Enter discount percentage: "))

# Discount multipliers for common percentages
multipliers = {0: 1.00, 5: 0.95, 10: 0.90, 15: 0.85, 20: 0.80, 25: 0.75, 30: 0.70}

# Fallback calculation for other percentages
multiplier = multipliers.get(discount_pct, (100 - discount_pct) / 100)
final_price = price * multiplier

print(f"Final price after discount: {final_price:.2f}")
```
