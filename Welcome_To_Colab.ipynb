```
#take a list of 15 numbers and ask the user to input any number and now search that element in the list  and remove all the occurences of that number from the list excluding the first occurence.
numbers = []

# Take 15 numbers as input
numbers = [int(input(f"Enter number {i+1}: ")) for i in range(15)]

# Input number to search
search_num = int(input("Enter number to search and modify: "))

# Process the list
if search_num in numbers:
    first_found = False
    for i in range(len(numbers)):
        if numbers[i] == search_num:
            if not first_found:
                first_found = True
            else:
                numbers[i] = None  # mark duplicates

    # Remove all marked elements
    numbers = [num for num in numbers if num is not None]

    print("Modified List:", numbers)
else:
    print("Number not found in the list")
```
