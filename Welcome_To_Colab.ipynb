```
#Question 14:
#Scenario:A security system reverses numeric codes.
#Task:Write a program to reverse a number.


# Taking input from user
num = int(input("Enter a number: "))

reverse = 0   # Variable to store reversed number

# Loop until number becomes 0
while num > 0:
    digit = num % 10          # Get last digit
    reverse = reverse * 10 + digit   # Add digit to reversed number
    num = num // 10           # Remove last digit

# Display result
print("Reversed number:", reverse)
```

```
#Question 15:
#Scenario:A system validates palindrome numbers.
#Task:Write a program to check if a number is palindrome.


# Taking input from user
num = int(input("Enter a number: "))

original = num   # Store original number
reverse = 0      # Variable to store reversed number

# Reverse the number
while num > 0:
    digit = num % 10           # Get last digit
    reverse = reverse * 10 + digit   # Build reversed number
    num = num // 10            # Remove last digit

# Check if original and reversed are same
if original == reverse:
    print("Palindrome Number")
else:
    print("Not a Palindrome Number")
```

```
#Question 16:
#Scenario:A math tool calculates factorial.
#Task:Write a program to compute factorial of a number.

# Program to calculate factorial of a number

# Taking input from user
num = int(input("Enter a number: "))

fact = 1   # Initialize factorial

# Check for negative number
if num < 0:
    print("Factorial not defined for negative numbers")
else:
    # Loop from 1 to num
    for i in range(1, num + 1):
        fact *= i   # Multiply each number

    # Display result
    print("Factorial of", num, "is:", fact)
```

```
#Question 19:
#Scenario:A validation tool counts digits.
#Task:Write a program to count number of digits.



# Taking input from user
num = int(input("Enter a number: "))

count = 0   # Variable to store digit count

# Handle case when number is 0
if num == 0:
    count = 1
else:
    # Convert negative number to positive
    num = abs(num)

    # Loop until number becomes 0
    while num > 0:
        num = num // 10   # Remove last digit
        count += 1        # Increase count

# Display result
print("Number of digits:", count)
```

```
#Question 22:
#Scenario:A validation tool checks palindrome strings.
#Task:Write a program to check if string is palindrome.

# Program to check if a string is palindrome

# Taking input from user
text = input("Enter a string: ")

# Convert to lowercase for case-insensitive comparison
text = text.lower()

# Reverse the string using slicing
reverse = text[::-1]

# Check if original and reversed are same
if text == reverse:
    print("Palindrome String")
else:
    print("Not a Palindrome String")
```

```
#Question 25:
#Scenario:A formatter converts text.
#Task:write a program to convert string to uppercase without built-in methods.

# Program to convert string to uppercase without using built-in functions

# Taking input from user
text = input("Enter a string: ")

result = ""   # Empty string to store result

# Loop through each character
for ch in text:

    # Check if character is lowercase letter
    if 'a' <= ch <= 'z':
        # Convert to uppercase using ASCII value
        upper_char = chr(ord(ch) - 32)
        result += upper_char
    else:
        # Keep character as it is (for digits, spaces, symbols)
        result += ch

# Display result
print("Uppercase string:", result)
```

```
#Question 26:
#Scenario:A parser extracts longest word.
#Task:Write a program to find longest word.


# Taking input from user
text = input("Enter a sentence: ")

# Split the sentence into words
words = text.split()

longest_word = ""   # Variable to store longest word

# Loop through each word
for word in words:

    # Check if current word is longer
    if len(word) > len(longest_word):
        longest_word = word   # Update longest word

# Display result
print("Longest word:", longest_word)
print("Length:", len(longest_word))
```

```
#Question 28:
#Scenario:A system checks anagrams.
#Task:Write a program to verify two strings are anagrams.



# Import Counter class for character frequency counting
from collections import Counter

# Input strings from user
str1 = input("Enter first string: ")
str2 = input("Enter second string: ")

# Count frequency of each character (case-insensitive)
count1 = Counter(str1.lower())  # Dictionary: {'l':1, 'i':1, 's':1, etc.}
count2 = Counter(str2.lower())  # Same format for second string

print("Character counts:")
print("String 1:", dict(count1))
print("String 2:", dict(count2))

# Compare frequency dictionaries directly
if count1 == count2:
    print(f"✓ '{str1}' and '{str2}' ARE anagrams!")
else:
    print(f"✗ '{str1}' and '{str2}' are NOT anagrams.")
```

```
#Question 30:
#Scenario:A data cleaner identifies duplicates.
#Task:Write a program to find duplicate characters.


# Input string
text = input("Enter a string: ")

# Dictionary to store frequency
freq = {}

# Count each character
for ch in text:
    if ch in freq:
        freq[ch] += 1
    else:
        freq[ch] = 1

# Print duplicate characters
print("Duplicate characters are:")
for ch in freq:
    if freq[ch] > 1 and ch != ' ':
        print(ch)
```

```
#Question 39:
#Scenario:A sequence is missing a number.
#Task:Write a program to find missing number.

# Input list
lst = list(map(int, input("Enter numbers: ").split()))

# Find range
n = len(lst) + 1   # because one number is missing

# Expected sum of numbers from 1 to n
expected_sum = n * (n + 1) // 2

# Actual sum of list
actual_sum = sum(lst)

# Missing number
missing = expected_sum - actual_sum

print("Missing number:", missing)
```

```
#Question 40:
#Scenario:A system rotates list data.
#Task:Write a program to rotate list by one position.

# Input list
lst = list(map(int, input("Enter elements: ").split()))

# Rotate right by one position
last = lst[-1]
for i in range(len(lst)-1, 0, -1):
    lst[i] = lst[i-1]
lst[0] = last

# Output result
print("List after rotation:", lst)
```

```
#Question 41:
#Scenario:A system stores student marks.
#Task:Write a program to find student with highest marks from dictionary.

# Input dictionary
n = int(input("Enter number of students: "))
students = {}

# Taking input
for i in range(n):
    name = input("Enter student name: ")
    marks = int(input("Enter marks: "))
    students[name] = marks

# Find student with highest marks
top_student = max(students, key=students.get)

# Output result
print("Student with highest marks:", top_student)
print("Marks:", students[top_student])
```

```
#Question 42:
#Scenario:A billing system calculates totals.
#Task:Write a function to calculate total bill amount.


# Function to calculate total bill
def calculate_bill(prices):
    total = 0
    for price in prices:
        total += price
    return total

# Input prices
prices = list(map(float, input("Enter item prices: ").split()))

# Function call
total_amount = calculate_bill(prices)

# Output result
print("Total bill amount:", total_amount)
```
