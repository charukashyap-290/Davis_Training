```
#write a program to print 20 sentences in dufferent lines.

sentences = []

# take input
for i in range(1, 21):
    s = input(f"Enter sentence {i}: ")
    sentences.append(s)

print("\n--- Your Sentences ---")

# Print im
for s in sentences:
    print(s)
```
