```
#To find hcf of 2 numbers
num1 = int (input("enter first number:"))
num2 = int (input("enter second number:"))
if (num2 == 0):
  print("HCF :" ,num1)
else:
   while(num2 !=0):
    rem = num1 % num2
    num1 = num2
    num2 = rem
    print("HCF :" ,num1)
```

```
#WAP to check a number is odd or even
def check_even_odd(num):
    if num % 2 == 0:
        return "Even number"
    else:
        return "Odd number"

# input from user
number = int(input("Enter a number: "))

# function call
result = check_even_odd(number)

# output
print(result)
```
