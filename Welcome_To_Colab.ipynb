```
#take a list of 15 numbers and ask the user to input any number and now search that element in the list  and remove all the occurences of that number from the list excluding the first occurence.
numbers = []

for i in range(15):
    numbers.append(int(input()))

search_num = int(input())

if search_num in numbers:
    new_list = []
    found = False

    for num in numbers:
        if num == search_num:
            if not found:
                new_list.append(num)
                found = True
        else:
            new_list.append(num)

    print(new_list)
else:
    print("Number not found")
```
