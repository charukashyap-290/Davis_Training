```
#Question 1: Find Duplicate Elements in a List
#Problem Statement:Write a program to find all duplicate elements in a given list.


lst = list(map(int, input("Enter elements separated by space: ").split()))

seen = set()
duplicates = set()

for num in lst:
    if num in seen:
        duplicates.add(num)
    else:
        seen.add(num)

print("Duplicate elements:", list(duplicates))

```

```
#Question 2: List Rotation
#Problem Statement:Rotate a list to the right by k positions.
# Rotate a list to the right by k positions

lst = list(map(int, input("Enter elements: ").split()))
k = int(input("Enter k: "))

n = len(lst)
k = k % n  # handle large k

rotated = lst[-k:] + lst[:-k]

print("Rotated List:", rotated)
```

```
#Question 3: Remove Common Elements from Two Lists
#Problem Statement:Remove elements from the first list that are present in the second list.


list1 = list(map(int, input("Enter elements of list1: ").split()))
list2 = list(map(int, input("Enter elements of list2: ").split()))

result = [x for x in list1 if x not in list2]

print("Result:", result)
```

```
#Question 4: Tuple Element Frequency
#Problem Statement:count the frequency of each element in a tuple.
# Tuple Element Frequency (simple, error-free)

t = tuple(input("Enter elements separated by space: ").split())

freq = {}

for item in t:
    freq[item] = freq.get(item, 0) + 1

print("Frequency:", freq)
```

```

#Question 5: Convert List to Tuple and Remove Duplicates
#Problem Statement:Convert a list into a tuple after removing duplicate elements.
# Convert list to tuple after removing duplicates

lst = list(map(int, input("Enter elements: ").split()))

unique = []

for i in lst:
    if i not in unique:
        unique.append(i)

result = tuple(unique)

print("Result:", result)
```

```
#Question 6: Find Maximum and Minimum in Tuple
#Problem Statement:Find the maximum and minimum elements in a tuple without using built-in functions.
# Find maximum and minimum in a tuple (without built-in functions)

t = tuple(map(int, input("Enter elements: ").split()))

max_val = t[0]
min_val = t[0]

for i in t:
    if i > max_val:
        max_val = i
    if i < min_val:
        min_val = i

print("Max =", max_val)
print("Min =", min_val)
```

```
#Question 7: Set Difference and Intersection
#Problem Statement:Find the intersection and difference between two sets.
# Set Difference and Intersection

set1 = set(map(int, input("Enter elements of set1: ").split()))
set2 = set(map(int, input("Enter elements of set2: ").split()))

intersection = set1 & set2
difference = set1 - set2

print("Intersection:", intersection)
print("Difference:", difference)
```

```
#Question 8: Check Subset
#Problem Statement:Check whether one set is a subset of another.


set1 = set(map(int, input("Enter elements of set1: ").split()))
set2 = set(map(int, input("Enter elements of set2: ").split()))

if set1.issubset(set2):
    print("True")
else:
    print("False")
```

```
#Question 9: Merge and Sort Unique Elements
#Problem Statement:Merge two lists and return a sorted list of unique elements.


list1 = list(map(int, input("Enter elements of list1: ").split()))
list2 = list(map(int, input("Enter elements of list2: ").split()))

merged = list1 + list2
unique = []

for i in merged:
    if i not in unique:
        unique.append(i)

unique.sort()

print("Result:", unique)
```

```
#Question 10: Tuple Pair Sum
#Problem Statement:Find all pairs in a tuple whose sum equals a given value.


t = tuple(map(int, input("Enter elements: ").split()))
target = int(input("Enter target sum: "))

pairs = []

for i in range(len(t)):
    for j in range(i + 1, len(t)):
        if t[i] + t[j] == target:
            pairs.append((t[i], t[j]))

print("Pairs:", pairs)
```

```
#Question 11: Remove Empty Sets from List
#Problem Statement:Remove all empty sets from a list of sets.
# Remove empty sets from a list

lst = [{1, 2}, set(), {3, 4}, set()]

result = []

for s in lst:
    if len(s) != 0:
        result.append(s)

print("Result:", result)
```
